```python
import pandas as pd
import nltk
from nltk import pos_tag, word_tokenize
```

```python
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
```

```python
def categorize_words(text):
    if pd.isnull(text):
        return []
    else:
        tokens = word_tokenize(text)
        pos_tags = pos_tag(tokens)
        for word,pos in pos_tags:
            if 'NNP' in pos:
                return "NNP"
            elif 'NNPS' in pos:
                return "NNP"
        return pos_tags
```
